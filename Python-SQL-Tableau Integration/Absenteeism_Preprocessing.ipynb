#### Lets get dummies for the 'Reason for Absence' column and store it in a separate dataframe.

```python
## set drop_first = True to avoid collinearity problem 

reason_columns = pd.get_dummies(df['Reason for Absence'], drop_first = True)
```

```python
reason_columns
```

#### we have to group the Reasons for Absence as we cannot use all the 28 dummies. From the Feature description file, the reasons 1 to 14 are related to various diseases, 15 to 17 are realted to pregnancy, 18 to 21 are related to poisoning and finally 22 to 28 are related to as the company calls lite reasons for absent. So, we group them as such.

```python
df.columns.values
```

```python
reason_columns.columns.values
```

```python
## lets drop the 'Reason for Absence' column from the dataframe before adding the dummy variables

df = df.drop(['Reason for Absence'], axis = 1)
```

```python
df.head()
```

```python
reason_type_1 = reason_columns.loc[:, 1:14].max(axis=1)
reason_type_2 = reason_columns.loc[:, 15:17].max(axis=1)
reason_type_3 = reason_columns.loc[:, 18:21].max(axis=1)
reason_type_4 = reason_columns.loc[:, 22:].max(axis=1)
```

```python
## lets concatenate these grouped dummies to the original dataframe

df = pd.concat([df, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis = 1)
df.head()
```

```python
df.columns.values
```

```python
column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
```

```python
df.columns = column_names
```

```python
df.head()
```

```python
## Lets reorder columns so the four dummy variables appears in the front 

column_names_reordered = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 
                          'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
df = df[column_names_reordered]
```

```python
df.head()
```

```python
## lets make a copy of this dataframe

df_reason_mod = df.copy()
```

```python
df_reason_mod.head()
```

#### Lets analyse the Date column:

```python
## initially the date column is defined as an object data type

type(df_reason_mod['Date'][0])
```

```python
## lets convert the date column to pandas 'datetime' type

df_reason_mod['Date'] = pd.to_datetime(df_reason_mod['Date'])
```

```python
df_reason_mod['Date']
```

```python
df_reason_mod['Date'] = pd.to_datetime(df_reason_mod['Date'], format = '%d/%m/%Y')
```

```python
df_reason_mod.info()
```

#### Extract the Month Value:

```python
df_reason_mod['Date'][0]
```

```python
df_reason_mod['Date'][0].month
```

```python
df_reason_mod.shape
```

```python
list_months = []

for i in range(df_reason_mod.shape[0]):
    list_months.append(df_reason_mod['Date'][i].month)
```

```python
list_months
```

```python
len(list_months)
```

```python
df_reason_mod['Month Value'] = list_months
```

```python
df_reason_mod.head(10)
```

#### Extract the Day of the Week:

```python
df_reason_mod['Date'][699].weekday()
```

```python
df_reason_mod['Date'][699]
```

```python
## function for extracting weekday from the date column

def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
df_reason_mod['Day of the Week'] = df_reason_mod['Date'].apply(date_to_weekday)
```

```python
df_reason_mod.head()
```

```python
## lets drop the date column as we extracted month and day of the week the employee was absent

df_reason_mod = df_reason_mod.drop(['Date'], axis = 1)
df_reason_mod.head()
```

```python
df_reason_mod.columns.values
```

```python
column_names_upd = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Month Value', 'Day of the Week',
       'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education', 'Children',
       'Pets', 'Absenteeism Time in Hours']
```

```python
## lets move the Month and Day of the week columns upfront where date column used to be

df_reason_mod = df_reason_mod[column_names_upd]
df_reason_mod.head()
```

```python
df_reason_date_mod = df_reason_mod.copy()
df_reason_date_mod.head()
```

#### Lets analyse the Education column:

```python
df_reason_date_mod['Education'].unique()
```

```python
## 1 represents 'High school',  2 for 'Graduate', 3 for 'Postgraduate' and 4 for 'a Master or a Doctor'

df_reason_date_mod['Education'].value_counts()
```

```python
## lets group all the Higher educations into one category

df_reason_date_mod['Education'] = df_reason_date_mod['Education'].map({1:0, 2:1, 3:1, 4:1})
```

```python
## lets make a final copy of the preprocessed dataframe

df_preprocessed = df_reason_date_mod.copy()
df_preprocessed.head()
```
